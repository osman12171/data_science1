```python
# df['zipcode']=df['zipcode'].astype('object')
# x=df[['sqft_living','grade','bathrooms','view','sqft_above','sqft_living15','sqft_basement','bedrooms']]
x=df[['sqft_living','grade','bathrooms','view','sqft_above','sqft_living15','sqft_basement','bedrooms','lat','waterfront','floors','yr_renovated','sqft_lot','yr_built','zipcode']]
y=df[['price']]
```

```python
# basari orani > 0.80 olmali ve
# rmse < $100K
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20, random_state=42) 

# x_train, x_test,y_train, y_test=train_test_split(x,y,test_size=0.20, random_state=42)
```

```python
lr=LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
tahmin=model.predict(x_test)
```

```python
r2_score(y_test,tahmin)
```
