```python
import numpy as np # lineer cebir, matrix, vektor uzaylari, 
import matplotlib.pyplot as plt # 
import pandas as pd # veri okuma ve manipulasyon
import seaborn as sns # fancy charts
```

```python
# fake data diye bisey var yani snythetic data. burada da Monte Carlo simulasyonu devreye giriyor, Asiye de ugrasiyor bunla. 
```

```python
# simdi x,y ve z de sayilar uretelim. 
```

```python
x=np.arange(20) # 1den 20 ye kadar ardisik rakamlar uret.
y=np.random.normal(10,1,20) # ortalama 10, standart sapma 1, rastgele 20 rakam uret.
z=np.random.normal(10,2,20) # ortalama 10, standart sapma 2, rastgele 20 rakam uret.
```

```python
plt.plot(x,y);  # burada islem sonuna ; koyunca grafigin ramdeki yerini gosteren yazilar gidiyor.
```

```python
plt.plot(x,y, label='y')
plt.plot(x,z, label='z')
plt.xlabel('x degerleri')
plt.ylabel('y degerleri')
plt.title('x ve y degerlerinin degisimi')
plt.legend();
```

```python
plt.scatter(x,y);
```

```python
fig=plt.figure(figsize=(10,6)) 
ax=fig.add_subplot(131) 
ax.plot(x,y,c='r',marker='d') # burada x ve y ana unsurlar. ondan sonra gelen hersey ekstra ozellikler. 

bx=fig.add_subplot(132)
bx.scatter(x,z,c='g',marker='*') # burada x ve z ana unsurlar. ondan sonra gelen hersey ekstra ozellikler.

cx=fig.add_subplot(133)
cx.hist(y); # hist ozelliginden dolayi sadece tek degisken alir. ancak cx.hist2d(x,y) yaparsak 2 degisken de alabilir.
```

#### Feature Engineering  Feature Engineering - Var olan veriden yeni sutunlar olusturma isine feature engineering diyoruz. Ozellik demek.

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['day']=(df['Date']).dt.day  # feature engineering yaptik.
df['month']=(df['Date']).dt.month  # feature engineering yaptik.
df['year']=(df['Date']).dt.year    # feature engineering yaptik.
```

```python
df.info()
```

```python
df.to_csv('supermarket_fe.csv') # bu yepyeni bir dosya olustu klasorde. yeni klasorun ya adini degistir kopyala ya da ustune yazdirilacagini bil.
```

```python
df['month_name']=(df['Date']).dt.month_name()
df['weekday']=(df['Date']).dt.day_name()
```
